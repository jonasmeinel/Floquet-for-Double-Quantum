```python
import qutip as q
import numpy as np
import subprocess
import ffmpeg
import pickle
import os
import re

%matplotlib inline
from matplotlib import rc
from matplotlib import pyplot as plt
font = {'weight' : 'normal',
        'size'   : 18}
rc('font', **font)

from matplotlib.font_manager import FontProperties
```

Our final goal is to solve understand this Hamiltonian for $\theta \neq 0$

$H = \frac{\omega_0}{2}  \,\sigma_z + \frac{\Omega_{rf}}{2}  \,cos(\omega_{rf}t) \,\left(\cos(\theta) \,\sigma_z + 2 \, \sin(\theta)\, \sigma_x\right) + \Omega_1 \, \cos(\omega_1 t) \, \sigma_x$

```python
#Frequency
Omega = 1
Omega *= 2 * np.pi

# NV hamiltonian
sz = q.sigmaz() * 0.5
sx = q.sigmax() * 0.5
sy = q.sigmay() * 0.5

# Comment: we adjust omega by sqrt(2) so we keep everything in spin 1/2 notation 

H0 = Omega * sx 

#time list
time_x=2
t = np.linspace(0, time_x, 500)

#initial state
a = 0
phi = np.cos(a/2.)*q.basis(2,0)+ (1 + 0j) * np.sin(a/2.)*q.basis(2,1)
rho_nv = phi*phi.dag()
```

### Rabi Oscillations

```python
resultat = []
resultat.append(q.mesolve(H=H0, rho0=rho_nv, tlist=t, c_ops = [], e_ops=[], options=q.Options(nsteps=100000)))


plt.plot(t,q.expect(resultat[0].states, sz))
```

# Time dependant Solutions (in lab frame !)

## Pure ODMR 

```python
#our TLS splitting
ws = 1e3
Hdc = ws * sz  

#our B1, and Brf (dressing fields)
B1 = 1
Brfz = 0
Brfx = 0

#and their operators
Hx1 = B1 * sx      #probe

Hdressing = Brfz * sz + Brfx * sx    #dressing field

#phases of B1 and dressing fields
phase_1 = 0
phase_d = 0


#time dependant functions
def Hx1_coeff(t,args):
    delta = args['wp'] 
    phase = args['phase']
    return np.sin(1*delta * t + phase)

def Hdressing_coeff(t,args):
    delta = args['dw'] 
    phase = args['phase']
    return on_off * np.sqrt(3) * np.cos( delta * t + phase)


#Overall Hamiltonian
H = [Hdc,[Hx1,Hx1_coeff],[Hdressing,Hdressing_coeff]]


#time array
time_x=2 * np.pi
t = np.linspace(0, time_x, 1000)




#initial state
a = 0
phi = np.cos(a/2.)*q.basis(2,0)+ (1 + 0j) * np.sin(a/2.)*q.basis(2,1)
rho_nv = phi*phi.dag()
print(rho_nv)
#

odmr_frq = []
odmr = []
l=41
wrf = 1
for k in range(l):
    wodmr = 0.3 * (k - l/2)
    wp = ws + wodmr
    odmr_frq.append(wodmr)
    output = q.mesolve(H=H, rho0=rho_nv, tlist=t, c_ops = [], e_ops=[], 
                       options=q.Options(nsteps=100000), args={'wp':wp ,'dw': wrf, 'phase': phase})
    odmr.append(q.expect(output.states, sz)[-1])
    
plt.plot(t,q.expect(output.states, sz))
```

```python
plt.figure(dpi = 100)

plt.title('ODMR')
plt.plot(odmr_frq,np.asarray(odmr)+0.5,marker = 'o', linestyle = '-')

plt.ylabel('$\\langle S_z \\rangle$')

comment = f'$\\omega_0 = {ws}$'
comment += '\n'
comment += f'$\\omega_x = {wrf}$'
comment += '\n'
comment += f'$\\Omega_x = {Brfx}$'


plt.text(-5, 0.6, comment, fontsize=14,
        verticalalignment='top')
plt.tight_layout()
plt.savefig('pure-odmr.png')
```

```python
plt.figure(dpi = 100)

plt.title('ODMR')
plt.plot(odmr_frq,np.asarray(odmr)+0.5,marker = 'o', linestyle = '-')

plt.ylabel('$\\langle S_z \\rangle$')

comment = f'$\\omega_0 = {ws}$'
comment += '\n'
comment += f'$\\omega_x = {wrf}$'
comment += '\n'
comment += f'$\\Omega_x = {Brfx}$'


plt.text(-6, 0.75, comment, fontsize=14,
        verticalalignment='top')
plt.tight_layout()
plt.savefig('dummy.png')
```
