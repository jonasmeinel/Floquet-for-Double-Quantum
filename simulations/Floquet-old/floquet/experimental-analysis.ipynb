```python
import numpy as np
from scipy import special
import matplotlib as mpl
import matplotlib.pyplot as plt



SMALL_SIZE = 20
MEDIUM_SIZE = 24
BIGGER_SIZE = 26

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rc('axes', linewidth=4)
mpl.rcParams['xtick.major.width'] = 4
mpl.rcParams['xtick.major.size'] = 10
mpl.rcParams['ytick.major.width'] = 4
mpl.rcParams['ytick.major.size'] = 10
#rc('xtick.major.width', linewidth=4)
```

```python
x = np.arange(0,10,0.1)
xx = 1.75
print('Omega = %.2f'%(xx * 1.4))

plt.figure(figsize=(7,3),dpi=130)
for k in range(3):
    plt.plot(x,special.jn(k,x), label = '$J_{%i}$'%k,linewidth=4)
    print('for x=%.2f, j%i(x) = '%(xx,k) + '%f.3'%special.jn(k,xx))
    print('Rabi period = %f.3' %(8/special.jn(k,xx) ))
plt.legend(loc=1)
plt.plot(np.ones(100) * xx, np.arange(100)/99.0 * 1.8 - 0.6,linewidth=4 )
plt.ylim([-0.6,1.2])
plt.ylabel('$P_n$')
plt.xlabel('$\\Omega_{\\rm{RF}}/\\omega_{\\rm{RF}}$')
plt.tight_layout()
plt.savefig('besselfunctions.png')
plt.show()

plt.plot(x,special.j1(x)/special.j0(x),'r',label = 'n=%i'%1)

plt.plot(x,special.jn(2,x)/special.j0(x),'b',label = 'n=%i'%2)
plt.plot(x,np.ones(len(x)) * 12/15.0,'c')
plt.plot(x,np.ones(len(x)) * 12/7.5,'c')
plt.plot(np.ones(len(x)) * xx, x,'m')
plt.ylim([0,3])
plt.xlim([0,2.5])
plt.legend()
plt.ylabel('$\\frac{J_n(x)}{ J_0(x)}$')
plt.xlabel('x')
plt.tight_layout()
plt.savefig('found_x_value.png')
plt.show()

print('x = %.2f'%xx)
```
